# Carrega CSV para análise e predição

```python
import pandas as pd
import numpy as np

import io
import requests
url = "https://raw.githubusercontent.com/Desnord/ProjetoFinalMC536/main/stage04/data/processed/analise.csv"
data = requests.get(url).content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
print(csv)
```

```python
from sklearn.utils import shuffle
csv = shuffle(csv)

y = csv['CASOS']
x = csv.drop('CASOS',1)
#x = pd.concat([pd.Series(1, index=x.index, name='x0'), x], axis=1)
print(x)
print(y)
```

```python
x = pd.get_dummies(x, prefix = ['ESTADO'], columns = ['ESTADO'])
print(x)
```

```python
x = np.split(x, [2808], axis=0)
x_teste   = x[0]
x_treino = x[1]

y = np.split(y, [2808], axis=0)
y_teste   = y[0]
y_treino = y[1]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# arvore de decisao para predicao de casos
lg = DecisionTreeRegressor(random_state=0)
lg.fit(x_treino, y_treino)

ypred = lg.predict(x_teste)

r2 = r2_score(y_teste,ypred)

print(r2)
```

```python
numsem = 0
numano = 0
qtdvoos = 0
ufestado = ''

siglas = est['UF'].tolist()

while numsem < 0 or numsem > 52 or ufestado not in siglas:
  print("Digite a semana: [0-52]")
  numsem = int(input())
  print("Digite o ano: ")
  numano = int(input())
  print("Digite a quantidade de voos no periodo: ")
  qtdvoos = int(input())
  print("Digite a sigla do estado: ")
  ufestado = input()

# encontra indice do estado digitado na lista
ind = 0
for i in range(len(siglas)):
  if siglas[i] == ufestado:
    ind = i 
    break

siglas = pd.Series(siglas)
siglas = pd.get_dummies(siglas,prefix='ESTADO')
siglas = siglas.values.tolist()

linha_predicao = [numsem] + [numano] + [qtdvoos] + siglas[ind]
valor = lg.predict([linha_predicao])
print("Numero de Casos Estimados: "+str(int(valor[0])))
```
